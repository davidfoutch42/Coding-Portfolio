```python
import networkx as nx
import pandas as pd
import sys
from PySide6.QtWidgets import QApplication, QDialog, QVBoxLayout, QWidget
from PySide6.QtWebEngineWidgets import QWebEngineView


class CentralityAnalysis:
    def __init__(self):
        super().__init__()
        # Initialize graph variable
        self.G = None
        
    def generate_graph(self):
        # Read data and create graph
        df = pd.read_csv('1YOK.cif_ringEdges', sep='\t')
        subset_df = df.loc[(df['NodeId1'].str.contains('A:')) & (df['NodeId2'].str.contains('A:'))]
        self.G = nx.from_pandas_edgelist(subset_df, 'NodeId1', 'NodeId2', create_using=nx.Graph())
        
    def generate_eigenVals(self):
        # Check if graph is generated
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        eigen_centr = nx.eigenvector_centrality_numpy(self.G)
        eigendf = pd.DataFrame(list(eigen_centr.items()), columns=['node', 'value'])
        eigendf['norm_val'] = eigendf['value'] / eigendf['value'].max()
        return eigendf
        
    def generate_closeVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        close_centr = nx.closeness_centrality(self.G)
        closedf = pd.DataFrame(list(close_centr.items()), columns=['node', 'value'])
        closedf['norm_val'] = closedf['value'] / closedf['value'].max()
        return closedf
        
    def generate_degrVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        degr_centr = nx.degree_centrality(self.G)
        degrdf = pd.DataFrame(list(degr_centr.items()), columns=['node', 'value'])
        degrdf['norm_val'] = degrdf['value'] / degrdf['value'].max()
        return degrdf
        
    def generate_betwVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        betw_centr = nx.betweenness_centrality(self.G)
        betwdf = pd.DataFrame(list(betw_centr.items()), columns=['node', 'value'])
        betwdf['norm_val'] = betwdf['value'] / betwdf['value'].max()
        return betwdf
            
    def generate_edgeBetw(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        edge_betw = nx.edge_betweenness_centrality(self.G)
        edge_betwdf = pd.DataFrame(list(edge_betw.items()), columns=['Pair', 'Edge Value'])
        edge_betwdf[['Node1', 'Node2']] = pd.DataFrame(edge_betwdf['Pair'].tolist(), index=edge_betwdf.index)
        edge_betwdf = edge_betwdf[['Node1', 'Node2', 'Edge Value']]
        edge_betwdf['Norm Value'] = edge_betwdf['Edge Value'] / edge_betwdf['Edge Value'].max()
        return edge_betwdf
    
class PlotlyGraph(QWidget):
    def __init__(self):
        super().__init__()

    def generate_plot(self, html_content):
        dialog = QDialog(self)
        dialog.setWindowTitle("Plot Window")
        dialog.setGeometry(200, 200, 800, 600)

        layout = QVBoxLayout(dialog)

        web_view = QWebEngineView()
        web_view.setHtml(html_content)
        layout.addWidget(web_view)

        dialog.setLayout(layout)
        dialog.exec_()

if __name__ == "__main__":
    app = QApplication(sys.argv)
    plotter = PlotlyGraph()

    # Instantiate CentralityAnalysis
    newGraph = CentralityAnalysis()
    newGraph.generate_graph()
    
    eigenvals = newGraph.generate_eigenVals()
    eigenvals['node'] = eigenvals['node'].str.replace(r'\D', '', regex=True)
    eigenvals['node'] = pd.to_numeric(eigenvals['node'])
    eigenvals = eigenvals.sort_values(by='node', ascending=True)
    eigenvals = eigenvals.drop(eigenvals.index[-1])
    
    betwvals = newGraph.generate_betwVals()
    betwvals['node'] = betwvals['node'].str.replace(r'\D', '', regex=True)
    betwvals['node'] = pd.to_numeric(betwvals['node'])
    betwvals = betwvals.sort_values(by='node', ascending=True)
    betwvals = betwvals.drop(betwvals.index[-1])

    degrvals = newGraph.generate_degrVals()
    degrvals['node'] = degrvals['node'].str.replace(r'\D', '', regex=True)
    degrvals['node'] = pd.to_numeric(degrvals['node'])
    degrvals = degrvals.sort_values(by='node', ascending=True)
    degrvals = degrvals.drop(degrvals.index[-1])

    closevals = newGraph.generate_closeVals()
    closevals['node'] = closevals['node'].str.replace(r'\D', '', regex=True)
    closevals['node'] = pd.to_numeric(closevals['node'])
    closevals = closevals.sort_values(by='node', ascending=True)
    closevals = closevals.drop(closevals.index[-1])

    # Your HTML content goes here
    line_plot_html = f'''
    <!DOCTYPE html>
    <html>
    <head>
        <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
        <style>
            .plot-div {{
                width: 800px;
                height: 300px;
            }}
        </style>
    </head>
    <body>
        <div id="plot1" class="plot-div"></div>
        <div id="plot2" class="plot-div"></div>
        <div id="plot3" class="plot-div"></div>
        <div id="plot4" class="plot-div"></div>
        <script>
            var trace1 = {{
                x: {list(eigenvals['node'])},
                y: {list(eigenvals['norm_val'])},
                mode: 'lines',
                type: 'scatter',
                name: 'Eigenvals',
                marker: {{ color: '#32CD32' }}
            }};
            var layout1 = {{
                title: 'Eigenvals Plot',
                xaxis: {{ title: 'Node' }},
                yaxis: {{ title: 'Norm Val', rangemode: 'tozero' }}
            }};
            Plotly.newPlot('plot1', [trace1], layout1);

            var trace2 = {{
                x: {list(betwvals['node'])},
                y: {list(betwvals['norm_val'])},
                mode: 'lines',
                type: 'scatter',
                name: 'Betwvals',
                marker: {{ color: '#FF5733' }}
            }};
            var layout2 = {{
                title: 'Betwvals Plot',
                xaxis: {{ title: 'Node' }},
                yaxis: {{ title: 'Norm Val', rangemode: 'tozero' }}
            }};
            Plotly.newPlot('plot2', [trace2], layout2);

            var trace3 = {{
                x: {list(degrvals['node'])},
                y: {list(degrvals['norm_val'])},
                mode: 'lines',
                type: 'scatter',
                name: 'Degrvals',
                marker: {{ color: '#0000FF' }}
            }};
            var layout3 = {{
                title: 'Degrvals Plot',
                xaxis: {{ title: 'Node' }},
                yaxis: {{ title: 'Norm Val', rangemode: 'tozero' }}
            }};
            Plotly.newPlot('plot3', [trace3], layout3);

            var trace4 = {{
                x: {list(closevals['node'])},
                y: {list(closevals['norm_val'])},
                mode: 'lines',
                type: 'scatter',
                name: 'Closevals',
                marker: {{ color: '#FFA500' }}
            }};
            var layout4 = {{
                title: 'Closevals Plot',
                xaxis: {{ title: 'Node' }},
                yaxis: {{ title: 'Norm Val', rangemode: 'tozero' }}
            }};
            Plotly.newPlot('plot4', [trace4], layout4);
        </script>
    </body>
    </html>
    '''

    plotter.generate_plot(line_plot_html)
    sys.exit(app.exec())
```

```python
import networkx as nx
import pandas as pd
import sys
import json  # Add this import for JSON serialization
from PySide6.QtWidgets import QApplication, QVBoxLayout, QWidget
from PySide6.QtWebEngineWidgets import QWebEngineView
from centrality_analysis_v2 import CentralityAnalysis
import pandas as pd

class CentralityAnalysis:
    def __init__(self):
        super().__init__()
        # Initialize graph variable
        self.G = None
        
    def generate_graph(self):
        # Read data and create graph
        df = pd.read_csv('1YOK.cif_ringEdges', sep='\t')
        subset_df = df.loc[(df['NodeId1'].str.contains('A:')) & (df['NodeId2'].str.contains('A:'))]
        self.G = nx.from_pandas_edgelist(subset_df, 'NodeId1', 'NodeId2', create_using=nx.Graph())
        
    def generate_eigenVals(self):
        # Check if graph is generated
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        eigen_centr = nx.eigenvector_centrality_numpy(self.G)
        eigendf = pd.DataFrame(list(eigen_centr.items()), columns=['node', 'value'])
        eigendf['norm_val'] = eigendf['value'] / eigendf['value'].max()
        return eigendf
        
    def generate_closeVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        close_centr = nx.closeness_centrality(self.G)
        closedf = pd.DataFrame(list(close_centr.items()), columns=['node', 'value'])
        closedf['norm_val'] = closedf['value'] / closedf['value'].max()
        return closedf
        
    def generate_degrVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        degr_centr = nx.degree_centrality(self.G)
        degrdf = pd.DataFrame(list(degr_centr.items()), columns=['node', 'value'])
        degrdf['norm_val'] = degrdf['value'] / degrdf['value'].max()
        return degrdf
        
    def generate_betwVals(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        betw_centr = nx.betweenness_centrality(self.G)
        betwdf = pd.DataFrame(list(betw_centr.items()), columns=['node', 'value'])
        betwdf['norm_val'] = betwdf['value'] / betwdf['value'].max()
        return betwdf
            
    def generate_edgeBetw(self):
        if self.G is None:
            raise ValueError("Graph not generated. Call generate_graph() first.")
        
        edge_betw = nx.edge_betweenness_centrality(self.G)
        edge_betwdf = pd.DataFrame(list(edge_betw.items()), columns=['Pair', 'Edge Value'])
        edge_betwdf[['Node1', 'Node2']] = pd.DataFrame(edge_betwdf['Pair'].tolist(), index=edge_betwdf.index)
        edge_betwdf = edge_betwdf[['Node1', 'Node2', 'Edge Value']]
        edge_betwdf['Norm Value'] = edge_betwdf['Edge Value'] / edge_betwdf['Edge Value'].max()
        return edge_betwdf

class PlotlyJSRegressionApp(QWidget):
    def __init__(self):
        super().__init__()

        layout = QVBoxLayout()

        # Create a QWebEngineView to display the Plotly plot
        self.webview = QWebEngineView()
        layout.addWidget(self.webview)

        self.setLayout(layout)
        
        newGraph = CentralityAnalysis()
        newGraph.generate_graph()

        eigenvals = newGraph.generate_eigenVals()
        eigenvals['node'] = eigenvals['node'].str.replace(r'\D', '', regex=True)
        eigenvals['node'] = pd.to_numeric(eigenvals['node'])
        eigenvals = eigenvals.sort_values(by='node', ascending=True)
        eigenvals = eigenvals.drop(eigenvals.index[-1])

        # Convert eigenvals to JSON string
        eigenvals_json = eigenvals.to_json(orient='records')
        
        betwvals = newGraph.generate_betwVals()
        betwvals['node'] = betwvals['node'].str.replace(r'\D', '', regex=True)
        betwvals['node'] = pd.to_numeric(betwvals['node'])
        betwvals = betwvals.sort_values(by='node', ascending=True)
        betwvals = betwvals.drop(betwvals.index[-1])
        
        # Convert eigenvals to JSON string
        betwvals_json = betwvals.to_json(orient='records')

        degrvals = newGraph.generate_degrVals()
        degrvals['node'] = degrvals['node'].str.replace(r'\D', '', regex=True)
        degrvals['node'] = pd.to_numeric(degrvals['node'])
        degrvals = degrvals.sort_values(by='node', ascending=True)
        degrvals = degrvals.drop(degrvals.index[-1])
        
        # Convert eigenvals to JSON string
        degrvals_json = degrvals.to_json(orient='records')

        closevals = newGraph.generate_closeVals()
        closevals['node'] = closevals['node'].str.replace(r'\D', '', regex=True)
        closevals['node'] = pd.to_numeric(closevals['node'])
        closevals = closevals.sort_values(by='node', ascending=True)
        closevals = closevals.drop(closevals.index[-1])
        
        # Convert eigenvals to JSON string
        closevals_json = closevals.to_json(orient='records')

        # HTML content with Plotly.js code and buttons to select datasets
        html_content = f'''
            <!DOCTYPE html>
            <html>
            <head>
                <title>Plotly Regression</title>
                <!-- Load Plotly.js library -->
                <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
                <style>
                    .selected {{
                        background-color: lightblue;
                    }}
                </style>
            </head>
            <body>
                <div id="plotlyDiv"></div>
                <div>
                    <!-- Buttons to select datasets -->
                    <button id="data1" onclick="changeData('1')">Data 1</button>
                    <button id="data2" onclick="changeData('2')">Data 2</button>
                    <button id="data3" onclick="changeData('3')">Data 3</button>
                    <button id="data4" onclick="changeData('4')">Data 4</button>
                    <button onclick="performRegression()">Plot Selected Data</button>
                </div>

                <script>
                    var eigenvals = JSON.parse('{eigenvals_json}');
                    var betwvals = JSON.parse('{betwvals_json}');
                    var degrvals = JSON.parse('{degrvals_json}');
                    var closevals = JSON.parse('{closevals_json}');

                    // Sample data for four datasets
                    var data1 = {{
                        x: eigenvals.map(item => item['node']),
                        y: eigenvals.map(item => item['norm_val']),
                        mode: 'markers',
                        name: 'Data 1'
                    }};

                    var data2 = {{
                        x: betwvals.map(item => item['node']),
                        y: betwvals.map(item => item['norm_val']),
                        mode: 'markers',
                        name: 'Data 2'
                    }};

                    var data3 = {{
                        x: degrvals.map(item => item['node']),
                        y: degrvals.map(item => item['norm_val']),
                        mode: 'markers',
                        name: 'Data 3'
                    }};

                    var data4 = {{
                        x: closevals.map(item => item['node']),
                        y: closevals.map(item => item['norm_val']),
                        mode: 'markers',
                        name: 'Data 4'
                    }};

                    var selectedData1 = null;
                    var selectedData2 = null;

                    // Function to highlight the selected button
                    function highlightButton(buttonId) {{
                        var buttons = document.getElementsByTagName('button');
                        for (var i = 0; i < buttons.length; i++) {{
                            buttons[i].classList.remove('selected');
                        }}
                        document.getElementById('data' + buttonId).classList.add('selected');
                    }}

                    // Create the plot using initial data
                    var layout = {{
                        title: 'Plotly Regression',
                        showlegend: true
                    }};

                    var plotlyDiv = document.getElementById('plotlyDiv');
                    Plotly.newPlot(plotlyDiv, [], layout);

                    // Function to change the selected datasets based on the button click
                    function changeData(dataNum) {{
                        highlightButton(dataNum);
                        var data = window['data' + dataNum];
                        if (selectedData1 === null) {{
                            selectedData1 = data;
                        }} else if (selectedData2 === null) {{
                            selectedData2 = data;
                        }} else {{
                            selectedData1 = data;
                            selectedData2 = null;
                        }}
                    }}

                    // Function to perform linear regression
                    function performRegression() {{
                        if (selectedData1 && selectedData2) {{
                            var x = selectedData1.y;
                            var y = selectedData2.y;

                            var trace1 = {{
                                x: x,
                                y: y,
                                mode: 'markers',
                                name: 'Data',
                                marker: {{
                                    color: 'rgba(255, 182, 193, .9)',
                                    size: 8
                                }}
                            }};

                            var regression = linearRegression(x, y);
                            var regressionLine = {{
                                x: x,
                                y: regression.y,
                                mode: 'lines',
                                name: 'Regression Line',
                                line: {{
                                    color: 'orange',
                                    width: 2
                                }}
                            }};

                            Plotly.react(plotlyDiv, [trace1, regressionLine], layout);
                        }}
                    }}

                    // Function to calculate linear regression
                    function linearRegression(x, y) {{
                        var slope, intercept, r;
                        var n = x.length;

                        var xMean = x.reduce((acc, val) => acc + val) / n;
                        var yMean = y.reduce((acc, val) => acc + val) / n;

                        var numerator = 0,
                            denominatorX = 0,
                            denominatorY = 0;

                        for (var i = 0; i < n; i++) {{
                            numerator += (x[i] - xMean) * (y[i] - yMean);
                            denominatorX += Math.pow(x[i] - xMean, 2);
                            denominatorY += Math.pow(y[i] - yMean, 2);
                        }}

                        slope = numerator / denominatorX;
                        intercept = yMean - slope * xMean;
                        r = numerator / Math.sqrt(denominatorX * denominatorY);

                        var regressionValues = [];
                        for (var i = 0; i < n; i++) {{
                            regressionValues.push(slope * x[i] + intercept);
                        }}

                        return {{ y: regressionValues, r: r }};
                    }}
                </script>
            </body>
            </html>
        '''

        # Load HTML content into QWebEngineView directly
        self.webview.setHtml(html_content)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = PlotlyJSRegressionApp()
    window.show()
    sys.exit(app.exec())
```
