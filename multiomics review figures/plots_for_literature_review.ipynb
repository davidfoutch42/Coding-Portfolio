```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6))
data = {'2012':11,'2013':9,'2014':12,'2015':21,'2016':24,'2017':16,'2018':36,'2019':37,'2020':30,'2021':59,'2022':13}
courses = list(data.keys())
values = list(data.values())
   
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.8)
 
plt.xlabel("Years in Review",fontsize=18)
plt.ylabel("Number of Articles",fontsize=18)
plt.title("Article Count by Year",fontsize=22)
plt.savefig('image.jpg')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(figsize=(12, 6))
    
X = ['2012','2013','2014','2015','2016','2017']
mRNA_microarray = [5,5,3,5,5,6]
miRNA_microarray = [0,3,2,1,2,1]
RNA_Seq = [2,3,0,1,8,4]
Clustering = [2,1,2,5,7,5]
Pathway_analysis = [5,4,5,11,12,9]
DEG = [6,3,3,7,10,8]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.25, mRNA_microarray, 0.09, label = ' mRNA_microarray')
plt.bar(X_axis - 0.15, miRNA_microarray, 0.09, label = 'miRNA_microarray')
plt.bar(X_axis - 0.05, RNA_Seq, 0.09, label = 'RNA_Seq')
plt.bar(X_axis + 0.05, Clustering, 0.09, label = 'Clustering')
plt.bar(X_axis + 0.15, Pathway_analysis, 0.09, label = 'Pathway_analysis')
plt.bar(X_axis + 0.25, DEG, 0.09, label = 'DEG')
plt.xticks(X_axis, X)
plt.xlabel("Genomics Methods by Year",fontsize=18)
plt.ylabel("Number of Articles", fontsize=18)
# plt.title("Number of Students in each group")
plt.legend()
plt.savefig('image2.jpg')
plt.show()
```
